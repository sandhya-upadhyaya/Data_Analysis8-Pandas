# 14. Check Chest Pain Type

```python
sns.countplot(x='cp', data = df)
plt.xticks([0,1,2,3],['typical angina', 'atypical angina','non-anginal pain','asymptimatic'] ,rotation=75 )
plt.show()
```

# 15. Show The Chest Pain Distribution As Per Target Variable

```python
sns.countplot(x='cp', hue='target', data = df)
plt.xticks([0,1,2,3],['typical angina', 'atypical angina', 'non-anginal pain', 'asymptimatic'], rotation= 75)
plt.legend(labels=['No-Disesses', 'Diseases'])
plt.show()
```

# 16. Show Fasting Blood Sugar Distribution According To Target Variable
